## polynomial regression


```python
## import libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
x=6 * np.random.rand(100,1)-3
y=0.5*x**2+1.5*x+2+np.random.randn(100,1)
## quadratic eqution used y=0.5x^2 +1.5x+2+outliers
plt.scatter (x,y,color='g')
plt.xlabel('X dataset')
plt.ylabel('y dataset')
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_trian,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
## let's implement linear regression 
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(x_train,y_trian
               )
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,regression.predict(x_train))
print(score)
```
